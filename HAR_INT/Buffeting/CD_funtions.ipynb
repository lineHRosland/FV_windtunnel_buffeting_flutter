```python
def fix(filtered, unfiltered):
    """
    Replace NaNs and their immediate neighbors in a filtered array with values from an unfiltered array.

    Parameters:
    - filtered: The array containing NaN values.
    - unfiltered: The original array with all valid values.

    Returns:
    - A new array where NaNs and their neighbors are replaced with values from the unfiltered array.
    """
    if filtered.shape != unfiltered.shape:
        raise ValueError("Filtered and unfiltered arrays must have the same shape.")

    result = filtered.copy()
    nan_indices = np.where(np.isnan(filtered))[0]
    indices_to_replace = set()

    for idx in nan_indices:
        indices_to_replace.add(idx)
        if idx > 0:
            indices_to_replace.add(idx - 1)
        if idx < len(filtered) - 1:
            indices_to_replace.add(idx + 1)

    for idx in indices_to_replace:
        result[idx] = unfiltered[idx]

    return result
```

```python
def poly_fit(x_MDS, x_MUS, y_MDS, y_MUS, plot_1=True, plot_2=True, plot_3=True, name=None, upwind=True, Save=False):
    import numpy as np
    import matplotlib.pyplot as plt
    from numpy.polynomial import Polynomial

    mask_MDS = (x_MDS >= -4) & (x_MDS <= 4)
    mask_MUS = (x_MUS >= -4) & (x_MUS <= 4)

    X_MDS = x_MDS[mask_MDS]
    Y_MDS = y_MDS[mask_MDS]
    X_MUS = x_MUS[mask_MUS]
    Y_MUS = y_MUS[mask_MUS]

    deg = 2
    p_MDS = Polynomial.fit(X_MDS, Y_MDS, deg).convert()
    p_MUS = Polynomial.fit(X_MUS, Y_MUS, deg).convert()

    derivative_MDS = p_MDS.deriv()
    derivative_MUS = p_MUS.deriv()

    Y_MDS_fit = p_MDS(X_MDS)
    Y_MUS_fit = p_MUS(X_MUS)

    average = (p_MDS(0) + p_MUS(0)) / 2
    Y_MDS_average = Y_MDS - p_MDS(0) + average
    Y_MUS_average = Y_MUS - p_MUS(0) + average
    Y_MDS_fit_average = Y_MDS_fit - p_MDS(0) + average
    Y_MUS_fit_average = Y_MUS_fit - p_MUS(0) + average

    cd_label = r'$C_{D,1}$' if upwind else r'$C_{D,2}$'
    x_label = r'$\alpha_1$ / $\alpha_2$ [deg]'
    label_MUS = f'{cd_label}($\\alpha_1$, 0)'
    label_MDS = f'{cd_label}(0, $\\alpha_2$)'

    if plot_1:
        fig1 = plt.figure(figsize=(10, 6))
        plt.plot(X_MDS, Y_MDS, label=label_MDS, color='blue')
        plt.plot(X_MUS, Y_MUS, label=label_MUS, color='red')
        plt.plot(X_MDS, Y_MDS_fit, '--', label=label_MDS + ' Fit', color='blue')
        plt.plot(X_MUS, Y_MUS_fit, '--', label=label_MUS + ' Fit', color='red')
        plt.xlabel(x_label, fontsize=14)
        plt.ylabel(cd_label, fontsize=14)
        #plt.title(name, fontsize=16)
        plt.grid(True, linestyle=':')
        plt.legend(fontsize=12)
        plt.tight_layout()
        if Save and name:
            fig1.savefig(f'{name}_plot1.png', dpi=300, bbox_inches='tight')
        plt.show()

    if plot_2:
        fig2 = plt.figure(figsize=(10, 6))
        plt.plot(X_MDS, Y_MDS_average, label=label_MDS, color='blue')
        plt.plot(X_MUS, Y_MUS_average, label=label_MUS, color='red')
        plt.plot(X_MDS, Y_MDS_fit_average, '--', label=label_MDS + ' Fit', color='blue')
        plt.plot(X_MUS, Y_MUS_fit_average, '--', label=label_MUS + ' Fit', color='red')
        plt.xlabel(x_label, fontsize=14)
        plt.ylabel(cd_label, fontsize=14)
        line1 = name if name else ''
        line2 = r'Aligned at $\alpha_1 = \alpha_2 = 0$'
        #plt.title(f'{line1}\n{line2}', fontsize=15)
        plt.grid(True, linestyle=':')
        plt.legend(fontsize=12)
        plt.tight_layout()
        if Save and name:
            fig2.savefig(f'{name}_plot2.png', dpi=300, bbox_inches='tight')
        plt.show()

    if plot_3:
        alpha1_vals = np.concatenate([np.zeros_like(X_MDS), X_MUS])
        alpha2_vals = np.concatenate([X_MDS, np.zeros_like(X_MUS)])
        y_vals = np.concatenate([p_MDS(X_MDS), p_MUS(X_MUS)])

        def poly2D_design_matrix(x, y, deg):
            terms = [(i, j) for i in range(deg + 1) for j in range(deg + 1 - i)]
            A = np.vstack([x**i * y**j for i, j in terms]).T
            return A, terms

        A, terms = poly2D_design_matrix(alpha1_vals, alpha2_vals, deg)
        coeffs, *_ = np.linalg.lstsq(A, y_vals, rcond=None)

        grid_x = np.linspace(-4, 4, 100)
        grid_y = np.linspace(-4, 4, 100)
        A1, A2 = np.meshgrid(grid_x, grid_y)
        Z = np.zeros_like(A1)
        for (i, j), c in zip(terms, coeffs):
            Z += c * A1**i * A2**j

        fig3 = plt.figure(figsize=(10, 6))
        ax = fig3.add_subplot(111, projection='3d')
        ax.plot_surface(A1, A2, Z, cmap='coolwarm', alpha=0.95)
        ax.set_xlabel(r'$\alpha_1$ [deg]', fontsize=12)
        ax.set_ylabel(r'$\alpha_2$ [deg]', fontsize=12)
        ax.set_zlabel(cd_label, fontsize=12)
        ax.zaxis.set_label_coords(-0.05, 0.5)
        line1 = name if name else ''
        line2 = r'2D polynomial surface fit from $\alpha_1$ and $\alpha_2$ data'
        #ax.set_title(f'{line1}\n{line2}', fontsize=14)
        fig3.subplots_adjust(right=0.85)
        ax.set_box_aspect(aspect=None, zoom=0.85)
        if Save and name:
            fig3.savefig(f'{name}_plot3.png', dpi=300, bbox_inches='tight')
        plt.show()

        x_eval = np.linspace(-3, 3, 1000)
        max_slope_MDS = np.max(np.abs(derivative_MDS(x_eval)))
        max_slope_MUS = np.max(np.abs(derivative_MUS(x_eval)))

        return average, max_slope_MDS, max_slope_MUS

```

```python
def CD(x_MDS, MDS_unfiltered, x_MUS, MUS_unfiltered, MDS_filtered = 0, MUS_filtered = 0, name = '', upwind = True):
    if np.all(MDS_filtered == 0):
        y_MDS = MDS_unfiltered
    else:
        y_MDS = fix(MDS_filtered, MDS_unfiltered)

    if np.all(MUS_filtered == 0):
        y_MUS = MUS_unfiltered
    else:
        y_MUS = fix(MUS_filtered, MUS_unfiltered)

    CD_mean, CD_d_MUS, CD_d_MDS = poly_fit(x_MDS, x_MUS, y_MDS, y_MUS, plot_1=True, plot_2=True, plot_3=True, name=name, upwind = upwind)

    filename = name + '.npy'
    save_dir = 'Static_coeff_for_buffeting'

    save_path = os.path.join(save_dir, filename)

    os.makedirs(save_dir, exist_ok=True)

    np.save(save_path, np.array([CD_mean, CD_d_MUS, CD_d_MDS]))

    print(f"Numpy array saved to {save_path}")

    print(CD_mean, CD_d_MUS, CD_d_MDS)
```
