```python
def fix(filtered, unfiltered):
    """
    Replace NaNs and their immediate neighbors in a filtered array with values from an unfiltered array.

    Parameters:
    - filtered: The array containing NaN values.
    - unfiltered: The original array with all valid values.

    Returns:
    - A new array where NaNs and their neighbors are replaced with values from the unfiltered array.
    """
    if filtered.shape != unfiltered.shape:
        raise ValueError("Filtered and unfiltered arrays must have the same shape.")

    result = filtered.copy()
    nan_indices = np.where(np.isnan(filtered))[0]
    indices_to_replace = set()

    for idx in nan_indices:
        indices_to_replace.add(idx)
        if idx > 0:
            indices_to_replace.add(idx - 1)
        if idx < len(filtered) - 1:
            indices_to_replace.add(idx + 1)

    for idx in indices_to_replace:
        result[idx] = unfiltered[idx]

    return result
```

```python
deg = 3
p1 = Polynomial.fit(x1_reduced, f1_reduced, deg).convert()
p2 = Polynomial.fit(x2_reduced, f2_reduced, deg).convert()
```

```python
# Evaluate fits
f1_fit = p1(x1_reduced)
f2_fit = p2(x2_reduced)

plt.figure(figsize=(10, 6))
plt.plot(x1_reduced, f1_reduced, label='f(a1, 0)', color='blue')
plt.plot(x1_reduced, f1_fit, '--', label='f(a1, 0) - Poly Fit', color='orange')
plt.plot(x2_reduced, f2_reduced, label='f(0, a2)', color='green')
plt.plot(x2_reduced, f2_fit, '--', label='f(0, a2) - Poly Fit', color='red')
plt.title("Polynomial Fit")
plt.xlabel("x1")
plt.ylabel("f")
plt.legend()
```

```python
def f_approx(x1, x2):
    return p1(x1) + p2(x2) - (p1(0)+p2(0))/2

X1, X2 = np.meshgrid(x1_reduced, x2_reduced)
F = f_approx(X1, X2)

x1_point = -3
x2_point = -3
z_point = f_approx(x1_point, x2_point)



fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X1, X2, F, cmap='viridis', alpha=0.8)
ax.scatter(
    x1_point, x2_point, z_point,
    color='red', s=100, marker='o',
    label='Point (-3, -3)',
    zorder=10
)

# Add the line with high zorder
ax.plot(
    [0, x1_point],
    [0, x2_point],
    [f_approx(0, 0), z_point],
    color='black',
    linewidth=2,
    linestyle='--',
    label='Line to Point',
    zorder=10
)

ax.set_title("2D Surface f(x1, x2) from Axis Functions")
ax.set_xlabel(r"$\alpha_1$")
ax.set_ylabel(r"$\alpha_2$")
ax.set_zlabel("f(x1, x2)")
ax.set_ylim(ax.get_ylim()[::-1]) 
plt.show()
```

```python
# Coordinates
x1_point = -3
x2_point = -3

# Z-values at origin and point
z_origin = f_approx(0, 0)
z_point = f_approx(x1_point, x2_point)

# Slopes in alpha_1 and alpha_2 directions
slope_alpha1 = (z_point - z_origin) / (x1_point)
slope_alpha2 = (z_point - z_origin) / (x2_point)

print(f"Slope in α1 direction: {slope_alpha1:.4f}")
print(f"Slope in α2 direction: {slope_alpha2:.4f}")
```

```python
# Derivatives of the fitted 1D polynomials
dp1 = p1.deriv()
dp2 = p2.deriv()

# Evaluate partial derivatives at x1 = -3 and x2 = -3
df_dalpha1 = dp1(x1_point)  # ∂f/∂α1 at (-3, -3)
df_dalpha2 = dp2(x2_point)  # ∂f/∂α2 at (-3, -3)

print(f"Partial derivative ∂f/∂α1 at (-3, -3): {df_dalpha1:.4f}")
print(f"Partial derivative ∂f/∂α2 at (-3, -3): {df_dalpha2:.4f}")
```
